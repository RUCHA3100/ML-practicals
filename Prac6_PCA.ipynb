```python
import pandas as pd
df=pd.read_csv("drug200.csv")
df.head()
```

```python
a1 =df["Age"]
a2 =df["Na_to_K"]

import matplotlib.pyplot as plt
plt.title("Age vs Na_to_K")
plt.xlabel("Age")
plt.ylabel("Na_to_K")
plt.plot(a1,a2,'.')
plt.show()
```

```python
import numpy as np
std1=(a1-np.mean(a1))/np.std(a1)
std2=(a2-np.mean(a2))/np.std(a2)
std3=np.array([std1.T,std2.T])

#covariance
covariance=np.cov(np.array([std1,std2]))

#EigenValue and EigenVector
from numpy import linalg as LA
eigen_val,eigen_vectors=LA.eig(covariance)
print(eigen_val)
print(eigen_vectors)
```

```python
new_feature=np.matmul(std3.T,eigen_vectors[1:].T)
new_feature[:10,:]
```

PCA using Scikitlearn Framework

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=1)
decomposed_pcs=pca.fit_transform(std3.T)
decomposed_pcs[:10,:]
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
decomposed_pcs=pca.fit_transform(std3.T)
decomposed_pcs
```
