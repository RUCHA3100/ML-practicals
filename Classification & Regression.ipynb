```python
import matplotlib.pyplot as plt
%matplotlib inline
 
# For creating dummy datasets using sklearn
from sklearn.datasets import make_regression
 
x,y = make_regression(n_samples=100, n_features=5, noise=50, random_state=42)
plt.scatter(x[:, 0],y)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
 
# For creating dummy datasets using sklearn
from sklearn.datasets import make_classification
 
x,y = make_classification(n_samples=100, n_features=5, n_redundant=1, n_informative=2, n_clusters_per_class=1, random_state=42)
plt.scatter(x[:, 0], x[:, 1], marker="o", c=y, s=25, edgecolor="k")
plt.show()
```
