```python
import pandas as pd
df=pd.read_csv("Credit Card Customer Data k-means.csv")
df.head()
```

```python
selected_data =df.iloc[:,2:4]
selected_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

k=3
kmeans=KMeans(k)
kmeans.fit(selected_data)
clusters=kmeans.fit_predict(selected_data)
clusters
```

```python
cluster_data=df.copy()
cluster_data['Clusters']=clusters
plt.scatter(cluster_data['Avg_Credit_Limit'],cluster_data['Total_Credit_Cards'],cmap='rainbow')
```

Dendrograms


```python
import pandas as pd
df=pd.read_csv("Credit Card Customer Data k-means.csv")
df.head()
```

```python
#normalize the data
from sklearn.preprocessing import normalize
norm_data=normalize(df)
norm_data=pd.DataFrame(norm_data,columns=df.columns)
norm_data.head()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title("Dendrograms of data")
dendrogram=shc.dendrogram(shc.linkage(norm_data,method='ward'))
```
